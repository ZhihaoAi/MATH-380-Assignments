```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
k1 = 0.2
k2 = 0.3
k3 = 0.001
k4 = 0.002
```

```python
O = lambda o, h: (1+k1)*o - k3*o*h
H = lambda o, h: (1+k2)*h - k4*o*h
```

```python
n = 25
o, h = 150, 200
os, hs = np.zeros(n), np.zeros(n)
ns = np.arange(n)
```

```python
def gen_list():
    os[0], hs[0] = o, h
    for i in range(n-1):
        os[i+1], hs[i+1] = O(os[i], hs[i]), H(os[i], hs[i])
```

```python
n = 25
o, h = 150, 200
os, hs = np.zeros(n), np.zeros(n)
ns = np.arange(n)
k1 = 0.19
k2 = 0.31
k3 = 0.001
k4 = 0.002
gen_list()
plt.scatter(ns, os, marker='o', label='Owls')
plt.scatter(ns, hs, marker='^', label='Hawks')
plt.legend(loc='upper left')
plt.xlabel('Days')
plt.ylabel('Birds')
plt.show()
```

```python
n = 25
m, o = 150, 200
ms, os = np.zeros(n), np.zeros(n)
ns = np.arange(n)
```

```python
M = lambda m, o: 1.2*m - 0.001*m*o
O = lambda m, o: 0.7*o + 0.002*m*o
```

```python
def gen_list_4():
    ms[0], os[0] = m, o
    for i in range(n-1):
        ms[i+1], os[i+1] = M(ms[i], os[i]), O(ms[i], os[i])
        if ms[i+1] < 0 or os[i+1] < 0:
            print(i)
            break
```

```python
n = 68
m, o = 200, 150
ms, os = np.zeros(n), np.zeros(n)
ns = np.arange(n)
gen_list_4()
plt.scatter(ns, ms, marker='^', label='Mice')
plt.scatter(ns, os, marker='o', label='Owls')
plt.legend(loc='upper left')
plt.xlabel('Years')
plt.ylabel('Population')
plt.show()
```

```python
import pandas as pd

hr = pd.read_csv('heart_rate.txt', header = None, sep='\t', names = ['body weight', 'pulse rate'])
#hr = hr.reset_index()
```

```python
hr['1/pulse rate'] = 1./hr['pulse rate']
```

```python
plt.scatter(x=hr['pulse rate'], y=hr['body weight'], marker='o')
```
